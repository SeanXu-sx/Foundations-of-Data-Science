```python
# Now that we know there are missing values in NumCompaniesWorked and TotalWorkingYears, let's check how many

df1.isnull().sum()
```

```python
# Yikes! 19 missing in NumCompaniesWorked and 9 missing in TotalWorkingYears
# Let's see how many records we have in total so we can determine if this is material

len(df1.index)
```

```python
# Less than 1% of data is missing from this sample size. Let's drop the rows where TotalWorkingYears is missing
# To ensure this change is permanent, remember to create a new object

df2 = df1.dropna(subset=['TotalWorkingYears'])
```

```python
# Let's check our new data frame. We can see we now have 9 records less

df2
```

```python
df2.isnull().sum()
```

```python
# Now let's deal with the the 19 missing data points. Let's assume that the numbers of companies someone has worked for is related to age.
# We should replace the NaN values with the mean number of companies worked for by age group.
# First, let's conduct a brief age analysis.

age_analysis = df2.groupby('Age')['NumCompaniesWorked'].agg(['count', 'mean']).reset_index()
age_analysis
```

```python
# We have a large range of ages! Perhaps we should create some bins so the data is easier to review

bins = [10,20,30,40,50,60,90]
bin_labels = ['10-19','20-29', '30-39', '40-49', '50-59', '>60']
df2['Age_Group'] = pd.cut(df2.Age, bins, right=False, labels = bin_labels)
```

```python
# If you scroll to the right, you should be able to see the new attribute

df2.head()
```

```python
# Alright. Let's try our age analysis again.

age_analysis = df2.groupby('Age_Group')['NumCompaniesWorked'].agg(['count', 'mean','sum']).reset_index()
age_analysis
```

```python
# What can we infer from the above?
# Well, the 'count' value tells us how many people we have in each bin, and the 'mean' tells us the avg # of companies they worked for
# This makes sense. Older people will have worked for more companies.
# Ok. Back to our missing data. Now that we have some mean values based on age bucket, we can use this to repalce NaN values

df2.loc[df2.NumCompaniesWorked.isnull(),'NumCompaniesWorked'] = df2.groupby('Age_Group').NumCompaniesWorked.transform('mean').round(1)
```

```python
# Now let's confirm how many NaNs we have...

df2.isnull().sum()
```

```python
# Success! No more NaNs. Now to simplify our analysis, is there any columns we feel we don't need?
# Since we already have age and age group, let's remove th "Over18" column

df2.drop(['Over18'],axis=1, inplace=True)
```

```python
# Let's confirm the columns again...note: this is a permanent drop of a column! You don't need a new object for this

list(df2.columns.values)
```

```python
#Let's pull some basic stats for all columns, incl the mean, standard deviation, and distribution of data

df2.describe()
```

```python
# Okay. Let's do some analysis - gender might be a good place to start
# We can use the crosstab function (this helps us pivot the data)

Gender_Analysis = pd.crosstab(df2['Age_Group'],df2['Gender'],df2['MonthlyIncome'],aggfunc='mean').round(1)
Gender_Analysis
```

```python
# Interesting. How about marital status?

Marital_Analysis = pd.crosstab(df2['MaritalStatus'],df2['Gender'],df2['MonthlyIncome'],aggfunc='mean').round(1)
Marital_Analysis
```

```python
# We can see that divorced women tend to make more than divorced men. I wonder how Job Level relates to this.

Level_Analysis = pd.crosstab(df2['MaritalStatus'],df2['JobLevel'],df2['MonthlyIncome'],aggfunc='mean').round(1)
Level_Analysis
```

```python
# Another interesting insight. Divorced people tend to make more than their married or single counterparts across most levels.
# Alright...let's try to visualize some of these insights.

Level_Analysis.plot(kind='bar',stacked=True)
```
